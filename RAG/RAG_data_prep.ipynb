```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split

pd.options.display.max_colwidth = 128
```

```python
attributes = pd.read_csv("/lakehouse/default/" + "Files/data/attributes.csv")
attributes
```

```python
attributes.dtypes
```

```python
attributes = attributes.dropna()
attributes
```

```python
descriptions = pd.read_csv("/lakehouse/default/" + "Files/data/product_descriptions.csv")
descriptions
```

```python
descriptions.iloc[7]['product_description']
```

```python
#https://wordlift.io/blog/en/gpt3-for-ecommerce/
#https://colab.research.google.com/drive/1zA2K0Dx_If5xPqGs2BE2dpjX_wMF7sqe?usp=sharing

# Add a space and dot when two words are not separated but skip abreviations
descriptions['product_description'] = descriptions['product_description'].str.replace("(?<=[a-z])(?=[A-Z])", r". ", regex=True)
# Some descriptions don't have a final dot at the end of the description
descriptions['product_description'] = descriptions['product_description'].str.rstrip('.') + '.'
# Add a space after the dot at the end of each sentence
descriptions['product_description'] = descriptions['product_description'].str.replace('.', '. ')
# Remove multiple spaces
descriptions['product_description'] = descriptions['product_description'].str.replace('  ',' ')
descriptions['product_description'] = descriptions['product_description'].str.strip()
```

```python
descriptions.iloc[7]['product_description']
```

```python
descriptions.iloc[8]['product_description']
```

```python
descriptions['remove_flag'] = descriptions['product_description'].apply(lambda x: 'California residents' in x)

descriptions = descriptions[descriptions['remove_flag'] == False]
descriptions = descriptions.drop(columns='remove_flag')
descriptions
```

```python
descriptions.dtypes
```

```python
descriptions = descriptions.dropna()
descriptions
```

```python
attributes['product_uid'] = attributes['product_uid'].astype('int64')
attributes.dtypes
```

```python
attributes['product_attributes'] = attributes['name'] + ': ' + attributes['value']
attributes
```

```python
attributes['product_attributes'] = attributes['product_attributes'].apply(lambda x: re.sub(r'\bBullet\d+: ', '', x))
attributes
```

```python
attributes_agg = attributes.groupby('product_uid').agg({'product_attributes': ' | '.join})
attributes_agg
```

```python
df_products = pd.merge(descriptions, attributes_agg, on='product_uid', how='inner')
df_products = df_products[['product_attributes', 'product_description']]
df_products
```

```python
df_products.to_csv("/lakehouse/default/" + "Files/data/df_products.csv", index=False)
```

```python
for i in range(10):
    print(df_products.iloc[i]['product_attributes'], '\n')
    print(df_products.iloc[i]['product_description'], '\n')
```

```python
df_products_train, df_products_test = train_test_split(df_products, test_size=500, random_state=369)
```

```python
print(len(df_products_train))
print(len(df_products_test))
```

```python
df_products_train.to_csv("/lakehouse/default/" + "Files/data/df_products_train.csv", index=False)
df_products_test.to_csv("/lakehouse/default/" + "Files/data/df_products_test.csv", index=False)
```
